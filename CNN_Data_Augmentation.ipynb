```python
# https://blog.keras.io/building-powerful-image-classification-models-using-very-little-data.html
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
datagen = ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest')
```

```python
# lets load the jpg image:
image=load_img("D:\Study\Python\scripts\Deep_Learning\CNN\Data_Augmentation/nawab.jpg")
```

```python
image
```

```python
x=img_to_array(image)
x=x.reshape((1,)+x.shape)          # changed 3X3 pixels into 4X4 pixels array
```

```python
# .flow() function of datagen basically generates numbers of random transformed images and save in the 'preview' directory:
```

```python
i=0
for batch in datagen.flow(x, batch_size=1, save_to_dir='D:\Study\Python\scripts\Deep_Learning\CNN\Data_Augmentation\Preview',save_prefix='Malik',save_format='jpeg'):
    i+=1
    if i>25:
        break          #otherwise it generates infinity images:
```
